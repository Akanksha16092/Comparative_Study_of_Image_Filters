```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Ensure original_image is available from previous steps
if 'original_image' not in locals() or original_image is None:
    print("Error: original_image is not loaded. Please ensure the image loading step was successful.")
else:
    h, w = original_image.shape[:2]

    # --- 1. Manual Translation ---
    tx, ty = 50, 30  # Translation amounts
    M_translate_manual = np.float32([[1, 0, tx], [0, 1, ty]])
    translated_image_manual = cv2.warpAffine(original_image, M_translate_manual, (w, h))

    # --- 2. Manual Rotation (45 degrees) ---
    center_x, center_y = w / 2, h / 2
    angle_rad = np.deg2rad(45) # Convert to radians
    cos_val = np.cos(angle_rad)
    sin_val = np.sin(angle_rad)

    # Rotation matrix around origin, then translate to center, rotate, translate back
    # Simplified: A standard rotation matrix for a point (x,y) around a center (cx,cy) is:
    # x' = cx + (x - cx) * cos(theta) - (y - cy) * sin(theta)
    # y' = cy + (x - cx) * sin(theta) + (y - cy) * cos(theta)
    # Which can be written in affine form:
    # M = [[cos(theta), -sin(theta), -cx*cos(theta) + cy*sin(theta) + cx],
    #      [sin(theta),  cos(theta), -cx*sin(theta) - cy*cos(theta) + cy]]
    M_rotate_manual = np.float32([
        [cos_val, -sin_val, -center_x * cos_val + center_y * sin_val + center_x],
        [sin_val,  cos_val, -center_x * sin_val - center_y * cos_val + center_y]
    ])
    rotated_image_manual = cv2.warpAffine(original_image, M_rotate_manual, (w, h))

    # --- 3. Manual Resizing (Smaller) ---
    scale_factor_smaller = 0.5
    M_scale_smaller_manual = np.float32([
        [scale_factor_smaller, 0, 0],
        [0, scale_factor_smaller, 0]
    ])
    new_w_smaller, new_h_smaller = int(w * scale_factor_smaller), int(h * scale_factor_smaller)
    resized_smaller_image_manual = cv2.warpAffine(original_image, M_scale_smaller_manual, (new_w_smaller, new_h_smaller))

    # --- 4. Manual Resizing (Larger) ---
    scale_factor_larger = 1.5
    M_scale_larger_manual = np.float32([
        [scale_factor_larger, 0, 0],
        [0, scale_factor_larger, 0]
    ])
    new_w_larger, new_h_larger = int(w * scale_factor_larger), int(h * scale_factor_larger)
    resized_larger_image_manual = cv2.warpAffine(original_image, M_scale_larger_manual, (new_w_larger, new_h_larger))

    # --- 5. Manual Horizontal Flipping ---
    # Matrix for horizontal flip: [[-1, 0, width-1], [0, 1, 0]]
    M_flip_h_manual = np.float32([[-1, 0, w - 1], [0, 1, 0]])
    flipped_h_image_manual = cv2.warpAffine(original_image, M_flip_h_manual, (w, h))

    # --- 6. Manual Vertical Flipping ---
    # Matrix for vertical flip: [[1, 0, 0], [0, -1, height-1]]
    M_flip_v_manual = np.float32([[1, 0, 0], [0, -1, h - 1]])
    flipped_v_image_manual = cv2.warpAffine(original_image, M_flip_v_manual, (w, h))

    # --- Display all images ---
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    axes = axes.flatten()

    images = {
        "Original Image": original_image,
        "Manual Translated Image": translated_image_manual,
        "Manual Rotated Image (45°)": rotated_image_manual,
        "Manual Resized Smaller": resized_smaller_image_manual,
        "Manual Resized Larger": resized_larger_image_manual,
        "Manual Horizontally Flipped": flipped_h_image_manual,
        "Manual Vertically Flipped": flipped_v_image_manual
    }

    for i, (title, img) in enumerate(images.items()):
        ax = axes[i]
        # Convert BGR to RGB for matplotlib display
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        ax.set_title(title)
        ax.axis('off')

    # Turn off any unused subplots
    for i in range(len(images), len(axes)):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

    print("All manual transformations using matrix multiplication applied and displayed.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a list of successfully loaded images and their names
images_to_process = []

if 'geometric_image' in locals() and geometric_image is not None:
    images_to_process.append((geometric_image, "Geometric Shape"))
if 'scene_image' in locals() and scene_image is not None:
    images_to_process.append((scene_image, "Natural Scene"))
if 'qr_code_image' in locals() and qr_code_image is not None:
    images_to_process.append((qr_code_image, "QR Code"))
if 'portrait_image' in locals() and portrait_image is not None:
    images_to_process.append((portrait_image, "Portrait"))

if not images_to_process:
    print("No images were loaded successfully. Please ensure previous steps loaded images correctly.")
else:
    print(f"Processing {len(images_to_process)} images.")

    # Iterate through each image and apply filters
    for original_image, image_name in images_to_process:
        print(f"\nApplying filters to: {image_name}")

        # Prepare filtered images list
        filtered_images = {
            "Original": original_image
        }

        # --- Gaussian Blur ---
        gaussian_sigmas = [0, 1, 2, 3]
        for sigma in gaussian_sigmas:
            # Kernel size (ksize) must be odd and positive. Using (5,5) as instructed.
            # sigmaX=0 means it's calculated from ksize, but task specifically asks for sigma=0,1,2,3.
            # If sigma is 0, OpenCV calculates it from ksize.
            # Let's keep ksize fixed as per instructions for consistency.
            blurred_image = cv2.GaussianBlur(original_image, (5, 5), sigmaX=sigma)
            filtered_images[f"Gaussian (Sigma={sigma})"] = blurred_image

        # --- Median Filter ---
        median_k_sizes = [3, 5]
        for k_size in median_k_sizes:
            median_image = cv2.medianBlur(original_image, k_size)
            filtered_images[f"Median (k={k_size})"] = median_image

        # --- Bilateral Filter ---
        # First set of parameters
        bilateral_image_1 = cv2.bilateralFilter(original_image, d=9, sigmaColor=75, sigmaSpace=75)
        filtered_images["Bilateral (d=9, sC=75, sS=75)"] = bilateral_image_1
        # Second set of parameters
        bilateral_image_2 = cv2.bilateralFilter(original_image, d=15, sigmaColor=150, sigmaSpace=150)
        filtered_images["Bilateral (d=15, sC=150, sS=150)"] = bilateral_image_2

        # --- Display results ---
        # Calculate grid size. 1 (Original) + 4 (Gaussian) + 2 (Median) + 2 (Bilateral) = 9 images
        # A 3x3 grid is suitable.
        num_filters = len(filtered_images) # Should be 9
        fig, axes = plt.subplots(3, 3, figsize=(18, 18))
        axes = axes.flatten()

        fig.suptitle(f"Filtering for: {image_name}", fontsize=16)

        for i, (title, img) in enumerate(filtered_images.items()):
            ax = axes[i]
            # Convert BGR to RGB for matplotlib display
            ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            ax.set_title(title)
            ax.axis('off')

        # Hide any unused subplots (though for 9 images, there won't be any if 3x3)
        for i in range(num_filters, len(axes)):
            fig.delaxes(axes[i])

        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
        plt.show()
    print("All requested filters applied and displayed for each image.")
```

## Add Noise to Images

### Subtask:
For each original image, generate two noisy versions: one with Gaussian noise and another with salt-and-pepper noise. Display the original, Gaussian noisy, and salt-and-pepper noisy versions side-by-side using subplots.


**Reasoning**:
I need to implement the noise generation functions as instructed and then apply them to each image in the `images_to_process` list, displaying the results in subplots for comparison.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ensure images_to_process is available from previous steps
# If this cell is run independently, re-create images_to_process (or ensure previous cells are run)
# For robustness, we'll redefine the noise functions as well.

def add_gaussian_noise(image, mean=0, std_dev=25):
    """
    Adds Gaussian noise to an image.
    """
    # Generate Gaussian noise
    gaussian_noise = np.random.normal(mean, std_dev, image.shape).astype(np.uint8)

    # Add noise to image
    noisy_image = cv2.add(image, gaussian_noise)

    # Clip pixel values to the valid range [0, 255]
    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
    return noisy_image

def add_salt_and_pepper_noise(image, probability=0.02):
    """
    Adds salt-and-pepper noise to an image.
    """
    noisy_image = np.copy(image)
    total_pixels = image.shape[0] * image.shape[1]

    # Add salt noise
    num_salt = np.ceil(probability * total_pixels * 0.5)
    coords_salt = [np.random.randint(0, s - 1, int(num_salt)) for s in image.shape[:2]]
    noisy_image[coords_salt[0], coords_salt[1], :] = 255 # Apply to all channels

    # Add pepper noise
    num_pepper = np.ceil(probability * total_pixels * 0.5)
    coords_pepper = [np.random.randint(0, s - 1, int(num_pepper)) for s in image.shape[:2]]
    noisy_image[coords_pepper[0], coords_pepper[1], :] = 0 # Apply to all channels

    return noisy_image

# Re-create images_to_process if not available (for independent execution safety)
if 'images_to_process' not in locals() or not images_to_process:
    # This block is for demonstrating safety; ideally, previous cells are run.
    # For this exercise, we assume `images_to_process` is populated from prior steps.
    print("Warning: 'images_to_process' not found or empty. Ensure previous image loading steps were successful.")
    # As a fallback for demonstration, you might load a single dummy image:
    # dummy_url = 'https://picsum.photos/200/200'
    # response = requests.get(dummy_url)
    # nparr = np.frombuffer(response.content, np.uint8)
    # dummy_image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    # images_to_process = [(dummy_image, "Dummy Image")] if dummy_image is not None else []

if not images_to_process:
    print("No images to process. Exiting.")
else:
    print(f"Processing {len(images_to_process)} images for smoothing noisy versions.")

    for original_image, image_name in images_to_process:
        print(f"\n--- Processing noisy versions of: {image_name} ---")

        # Generate noisy images
        gaussian_noisy_image = add_gaussian_noise(original_image, mean=0, std_dev=25)
        salt_pepper_noisy_image = add_salt_and_pepper_noise(original_image, probability=0.02)

        # List of filter parameters to apply
        gaussian_sigmas = [0, 1, 2, 3]
        median_k_sizes = [3, 5]
        bilateral_params = [
            {"d": 9, "sigmaColor": 75, "sigmaSpace": 75},
            {"d": 15, "sigmaColor": 150, "sigmaSpace": 150}
        ]

        # --- Apply filters to Gaussian noisy image ---
        print(f"Applying filters to Gaussian noisy {image_name}...")
        filtered_gaussian_noisy_images = {
            "Gaussian Noisy": gaussian_noisy_image
        }

        for sigma in gaussian_sigmas:
            blurred_image = cv2.GaussianBlur(gaussian_noisy_image, (5, 5), sigmaX=sigma)
            filtered_gaussian_noisy_images[f"Gaussian (Sigma={sigma})"] = blurred_image

        for k_size in median_k_sizes:
            median_image = cv2.medianBlur(gaussian_noisy_image, k_size)
            filtered_gaussian_noisy_images[f"Median (k={k_size})"] = median_image

        for params in bilateral_params:
            bilateral_image = cv2.bilateralFilter(
                gaussian_noisy_image, d=params["d"], sigmaColor=params["sigmaColor"], sigmaSpace=params["sigmaSpace"]
            )
            filtered_gaussian_noisy_images[f"Bilateral (d={params['d']}, sC={params['sigmaColor']})"] = bilateral_image

        # Display filtered Gaussian noisy images
        num_filters = len(filtered_gaussian_noisy_images)
        fig_gauss, axes_gauss = plt.subplots(3, 3, figsize=(18, 18))
        axes_gauss = axes_gauss.flatten()

        fig_gauss.suptitle(f"Smoothing Filters on Gaussian Noisy: {image_name}", fontsize=16)

        for i, (title, img) in enumerate(filtered_gaussian_noisy_images.items()):
            ax = axes_gauss[i]
            ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            ax.set_title(title)
            ax.axis('off')

        for i in range(num_filters, len(axes_gauss)): # Hide unused subplots
            fig_gauss.delaxes(axes_gauss[i])

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

        # --- Apply filters to Salt-and-Pepper noisy image ---
        print(f"Applying filters to Salt-and-Pepper noisy {image_name}...")
        filtered_sp_noisy_images = {
            "S&P Noisy": salt_pepper_noisy_image
        }

        for sigma in gaussian_sigmas:
            blurred_image = cv2.GaussianBlur(salt_pepper_noisy_image, (5, 5), sigmaX=sigma)
            filtered_sp_noisy_images[f"Gaussian (Sigma={sigma})"] = blurred_image

        for k_size in median_k_sizes:
            median_image = cv2.medianBlur(salt_pepper_noisy_image, k_size)
            filtered_sp_noisy_images[f"Median (k={k_size})"] = median_image

        for params in bilateral_params:
            bilateral_image = cv2.bilateralFilter(
                salt_pepper_noisy_image, d=params["d"], sigmaColor=params["sigmaColor"], sigmaSpace=params["sigmaSpace"]
            )
            filtered_sp_noisy_images[f"Bilateral (d={params['d']}, sC={params['sigmaColor']})"] = bilateral_image

        # Display filtered Salt-and-Pepper noisy images
        num_filters = len(filtered_sp_noisy_images)
        fig_sp, axes_sp = plt.subplots(3, 3, figsize=(18, 18))
        axes_sp = axes_sp.flatten()

        fig_sp.suptitle(f"Smoothing Filters on Salt-and-Pepper Noisy: {image_name}", fontsize=16)

        for i, (title, img) in enumerate(filtered_sp_noisy_images.items()):
            ax = axes_sp[i]
            ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            ax.set_title(title)
            ax.axis('off')

        for i in range(num_filters, len(axes_sp)): # Hide unused subplots
            fig_sp.delaxes(axes_sp[i])

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

    print("All smoothing filters applied to noisy images and displayed.")
```
